```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
# x: Number of floors of a building 
# Y: Annual Electric bill (USD)

x_train=np.array([5,10,15,20,25,30])
y_train=np.array([20,48,67,98,120,138])
x_test=np.array([])
y_test=np.array([])
print("x_train :",x_train)
print("y_train :",y_train)
```

```python
import random

# Generate the random values in an array
m = [random.randint(-20, 20) for i in range(1000)]
b = [random.randint(-20, 20) for i in range(1000)]
```

```python
plt.scatter(m, b,c='b')
plt.title("Parameters")
plt.ylabel('m')
plt.xlabel('b')
```

```python
# Plot the data points
plt.scatter(x_train, y_train,c='b')
plt.title("Electric charge")
plt.ylabel('Electric bill(USD)')
plt.xlabel('Number of devices')
```

```python
# defined Y^ (predcition value) for given m,b
def compute_model_prediction(x, m, b):
    w = x.shape[0]
    a = np.zeros(w)
    for i in range(w):
        a[i] = m * x[i] + b
        
    return a
```

```python
def compute_cost(x, y, m, b): 
    w = x.shape[0] 
    
    c_sum = 0 
    for i in range(w): 
        a = m * x[i] + b   
        cost = (a - y[i]) ** 2  
        c_sum = c_sum + cost  
    total_cost = (1 / (2 * w)) * c_sum  

    return total_cost
```

```python
temp_m=[]
for i in m:
  temp_m.append(compute_model_prediction(x_train, i, 0))
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='b',label='Actual Values')
plt.scatter(x_train, y_train,c='b')
plt.title("Electric charge")
plt.ylabel('Electric bill(USD)')
plt.xlabel('Number of floors')


# plot the prediction
for i in range(len(temp_m)):
  plt.plot(x_train, temp_m[i], c='r',label='Prediction')                     
                  
plt.show()  
```

```python
# cost funtion for each m
temp_jwb=[]
for i in m:
  temp_jwb.append(compute_cost(x_train, y_train, i, 0))
```

```python
plt.plot(m,temp_jwb,c='b')
plt.title("cost funtion")
plt.ylabel('j(m)')
plt.xlabel('m')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(m, b, temp_jwb , c='r', marker='o')
ax.set_xlabel('m')

ax.set_ylabel('b')
ax.set_zlabel('temp_jwb')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [7.50, 20.0]
plt.rcParams["figure.autolayout"] = True

plt.plot(m, b, 'r*')
plt.axis([0, 6, 0, 20])

for i, j,k in zip(m, b,temp_jwb):
  if k==min(temp_jwb):
    plt.text(i, j+0.5, '({}, {}, {})'.format(i, j,k))

plt.show()
```

```python
bOf_fwb=[]
for i,j,k in zip(m, b,temp_jwb):
  if k==min(temp_jwb):
    bOf_fwb.append(j)
print(bOf_fwb)
```

```python
final_m=5
cost=[]
for b in bOf_fwb:
    cost.append(compute_cost(x_train, y_train,final_m, b))
print('cost:',cost)

#min cost => good accuracy

final_b=bOf_fwb[cost.index(min(cost))]

print("Perameter b :",final_b)
```

```python
def model(x):
    y=final_m*x+final_b
    #print("Y :",y)
    return y
```
